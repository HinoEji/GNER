## Import Requirements

```python
import nltk
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, AutoModelForSeq2SeqLM
```

## Demo Input and Instructions

```python
# Demo input
sentence = "did george clooney make a musical in the 1980s"
words = nltk.word_tokenize(sentence)
entity_labels = ["genre", "rating", "review", "plot", "song", "average ratings", "director", "character", "trailer", "year", "actor", "title"]
# print(demo_words)

# fit in the instruction template
instruction_template = "Please analyze the sentence provided, identifying the type of entity for each word on a token-by-token basis.\nOutput format is: word_1(label_1), word_2(label_2), ...\nWe'll use the BIO-format to label the entities, where:\n1. B- (Begin) indicates the start of a named entity.\n2. I- (Inside) is used for words within a named entity but are not the first word.\n3. O (Outside) denotes words that are not part of a named entity.\n"
instruction = f"{instruction_template}\nUse the specific entity tags: {', '.join(entity_labels)} and O.\nSentence: {' '.join(words)}"
print(f"Final Instruction:\n\n{instruction}")
```

```python
# Demo input
sentence = "Tiki giao hàng nhanh"
words = sentence.split()
entity_labels = ["BRANDING", "DELIVERY", "GENERAL", "PRICE", "PRODUCT", "PRODUCT DESIGN", "PRODUCT FEATURE", "PRODUCT QUALITY", "PRODUCT USAGE", "SERVICE"]
# print(demo_words)

# fit in the instruction template
instruction_template = "Hãy phân tích câu được cung cấp, xác định loại thực thể cho mỗi từ.\nĐịnh dạng đầu ra: từ_1(nhãn_1), từ_2(nhãn_2), ...\nChúng ta sử dụng định dạng BIO để gán nhãn thực thể, trong đó:\n1. B- (Begin) chỉ ra sự bắt đầu của một thực thể có tên.\n2. I- (Inside) được sử dụng cho các từ trong một thực thể nhưng không phải từ đầu tiên.\n3. O (Outside) biểu thị các từ không phải là một phần của thực thể có tên.\n"
instruction = f"{instruction_template}\nSử dụng các nhãn thực thể sau: {', '.join(entity_labels)} và O.\nCâu: {' '.join(words)}"
print(f"Final Instruction:\n\n{instruction}")
with torch.inference_mode():
    inputs = tokenizer(instruction, return_tensors="pt").to("cuda")
    outputs = model.generate(**inputs, max_new_tokens=640, num_beams=5, do_sample=True)
    response = tokenizer.decode(outputs[0], skip_special_tokens=True).strip()
    print(f"Model Generation: \n\n{response}")
```

## Structure

```python
# structure the generated text
from evaluate import extract_predictions, parser
example = {
    "label_list": entity_labels,
    "instance": {"words": words},
    "prediction": response,
}
```

```python
# bio-format prediction
bio_predictions = extract_predictions(example)
print(f"Predictions (BIO-format): \n\n{bio_predictions}")
```

```python
# entity-level prediction
entity_level_predictions = parser(words, bio_predictions)
print(f"Predictions (Entity-level): \n\n{entity_level_predictions}")
```

# calcualte new metric

```python
from seqeval.metrics.sequence_labeling import get_entities
from evaluate import extract_predictions, parser
import json

def read_jsonl(path):
    data = []
  
    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue

            obj = json.loads(line)
            token = obj['instance']['words']
            response = obj['prediction']
            bio_label = obj['instance']['labels']
            label_bounds = get_entities(bio_label)

            pred = {
                "label_list": obj['label_list'],
                "instance": {"words": token},
                "prediction": response,
            }
            bio_pred = extract_predictions(pred, tokenizer)
            entity_pred = parser(token, bio_pred)
            entity_label = parser(token, bio_label)

            data.append(
                {
                    "tokens": token,
                    "bio_labels" : bio_label,
                    "entity_labels": entity_label,
                    "bio_preds" : bio_pred,
                    "entity_preds" : entity_pred,
                    "response" : response,
                    "list_entities" : [x.lower() for x in obj['label_list']]
                }
            )


    return data

from collections import defaultdict
from seqeval.metrics import classification_report as type_classification_report
from sklearn.metrics import classification_report as tag_classification_report
import numpy as np
def compute_metrics_by_entity_type(dataset):
    # list_entities = dataset[0]['list_entities']
    # metrics = {
    #     _type : {"n_correct" : 0, "n_pos_gold" : 0, "n_pos_pred" : 0} for _type in list_entities
    # }

    # for sample in dataset:
    #     for pred in sample["entity_preds"]:
    #         if pred in sample["entity_labels"]:
    #             metrics[pred[-1]]["n_correct"] +=1
    #         metrics[pred[-1]]["n_pos_pred"] +=1
    #     for gold in sample["entity_labels"]:
    #         metrics[gold[-1]]["n_pos_gold"] +=1
    # result = {}

    # for key,value in metrics.items():
    #     prec = value['n_correct'] / (value['n_pos_pred'] + 1e-10)
    #     recall = value['n_correct'] / (value['n_pos_gold'] + 1e-10)
    #     f1 = 2 * prec * recall / (prec + recall + 1e-10)
    #     result[key] = {"recall": recall, "precision" : prec, "f1":f1}
    # return result
    y_true = []
    y_pred = []
    y_true_flat = []
    y_pred_flat = []
    for sample in dataset:
        y_true.append(sample["bio_labels"])
        y_pred.append(sample["bio_preds"])

        y_true_flat.extend(sample["bio_labels"])
        y_pred_flat.extend(sample["bio_preds"])
    
    metrics_by_type = type_classification_report(y_true, y_pred, digits=4,output_dict=True)
    metrics_by_type = {k: int(v) if isinstance(v, np.int64) else v for k, v in metrics_by_type.items()}
    metrics_by_tag = tag_classification_report(y_true_flat, y_pred_flat, digits=4,output_dict=True)
    metrics_by_tag = {k: int(v) if isinstance(v, np.int64) else v for k, v in metrics_by_tag.items()}

    return {
        "metrics_by_type": metrics_by_type,
        "metrics_by_tag" : metrics_by_tag
    }

    
```

```python
samples = read_jsonl("/kaggle/working/GNER/output/vit5-base-10125-task-adaptation-beam1/predict_text_generations.jsonl")
```

```python
metrics = compute_metrics_by_entity_type(samples)
```
